```python
# 13)Which brands of vehicles are less affected by cost depreciation ?
brand_depreciation = df.groupby('Car_Name')['Depreciation'].mean().sort_values()
print("These are brands of vehicles are less affected by cost depreciation:")
print(brand_depreciation.head(6).index.tolist())
print("\n")
plt.figure(figsize=(14, 7))
sns.barplot(x=brand_depreciation.index, y=brand_depreciation.values)
plt.xticks(rotation=90)
plt.title('Average Depreciation Rate by Vehicle Brand')
plt.xlabel('Brand')
plt.ylabel('Average Depreciation Rate')
plt.tight_layout()
plt.show()
```

```python
# 14)Are there any factors which you feel affect the cost depreciation ?
numerical_columns = ['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Owner', 'Depreciation']
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heat Map')
plt.show()
```

From this heat map, it is clear that Present_Price and Selling_Price are the most significant factors affecting depreciation, followed by Kms_Driven and Year. The number of previous owners has a lesser impact on depreciation.

7*Analysis Categorical Variables*

```python
cat_cols = ['Year','Fuel_Type','Seller_Type','Transmission','Owner']
```

```python
plt.figure(figsize = [20,10])
i=1
for col in cat_cols:
        plt.subplot(2,3,i)
        sns.barplot(x=df[col],y=df['Depreciation'])
        plt.title(f'{col} vs Depreciation\n', fontdict={'size':25})
        plt.xlabel(col, size=15)
        plt.xticks(size=15, rotation=90)
        i+=1


plt.tight_layout()
plt.show()
```

**Insights**

*   Older cars (from years 2003 to 2009) generally show higher depreciation compared to newer cars (from years 2010 onwards).
*   Diesel cars tend to have higher depreciation compared to Petrol and CNG cars. And CNG cars have lowest depreciation.
*   Cars sold by Dealers show higher depreciation compared to those sold by Individuals.
*   Automatic transmission cars show higher depreciation compared to Manual transmission cars.
*   The number of owners significantly impacts the depreciation value, with higher ownership leading to higher depreciation.







*Analysis Numerical Variables*

```python
num_cols = ['Selling_Price','Kms_Driven','Present_Price','Depreciation']
```

```python
sns.pairplot(df[num_cols])
plt.show()
```

**Insights**

*   The histogram of the selling price is right-skewed, indicating that most cars have a lower selling price, with fewer cars having a high selling price.

*   There is a weak negative correlation between kilometers driven and selling price, suggesting that cars with higher mileage tend to have lower selling prices.


*   There is a strong positive correlation between the present price and the selling price. Higher present prices tend to result in higher selling prices.


*   There is a positive correlation between depreciation and selling price, indicating that cars with higher depreciation also tend to have higher selling prices. This might be due to the fact that more expensive cars (higher present price) depreciate more in absolute terms but still have higher selling prices.


*   There is a weak negative correlation between kilometers driven and the present price, indicating that cars with higher mileage generally have a lower present price.

*   There is a slight positive correlation between depreciation and kilometers driven, suggesting that cars that have been driven more have higher depreciation.


*   There is a strong positive correlation between the present price and depreciation. Cars with higher present prices tend to have higher depreciation.
